```python
import pandas as pd

data = pd.read_csv('pm10_data.csv', encoding='cp949')
df = pd.DataFrame(data)
```

```python
# column명에서 '월' 제거
df.columns = df.columns.str.replace('월', '').str.strip()

# 쓸모 없는 column drop
df = df.drop(columns=['항목', '단위'])

# 가로형 데이터를 세로형(표준 형태)으로 변환
df_long = df.melt(id_vars=['구분'],
                  var_name='YearMonth',
                  value_name='PM10')

df_long.head()
```

```python
# 날짜 형식을 분기(Quarter)로 변환
df_long['Quarter'] = pd.to_datetime(df_long['YearMonth'], format='%Y.%m', errors='coerce').dt.to_period('Q')

# 지역 및 분기별 집계
df_long['PM10'] = df_long['PM10'].astype(str).str.replace(r'\*+', '', regex=True)
df_long['PM10'] = pd.to_numeric(df_long['PM10'], errors='coerce')

df_quarterly = df_long.groupby(['구분', 'Quarter'])['PM10'].mean().reset_index()
df_quarterly['PM10'] = df_quarterly['PM10'].round(2)
df_quarterly.head()
```
